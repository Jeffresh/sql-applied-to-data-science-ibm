## Which food item has maximum sodium content?

- Main source of sodium is table salt
- Average American eats 5 teaspoons/day


```python
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns

### Categorical scatterplots 

plot = sns.swarmmplot(x="Category", y="Sodium", data=df)
plt.setp(plot.get_xticklabels(), rotation=70)
plt.title("Sodium Content')
plt.show()
```

Firstly we use data visualization to explore the sodium content of food items:

- We create a categorical scatterplot

```python
df['Sodium'].describe()
```
