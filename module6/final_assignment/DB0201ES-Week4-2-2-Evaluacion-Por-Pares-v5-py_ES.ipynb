## Problemas

Escribe y ejecuta las consultas SQL para resolver los problemas del trabajo

### Problema 1

##### Encuentra el numero total de crímenes almacenados en la tabla CRIME


```python
# Filas de la Tabla Crime
%sql SELECT COUNT(*) AS N_TOTAL_CRIMES FROM CHICAGO_CRIME_DATA
```

### Problema 2

##### Obtén las primeras 10 filas de la tabla CRIME


```python
%sql SELECT * FROM CHICAGO_CRIME_DATA LIMIT 10
```

### Problema 3

##### ¿Cuántos delitos involucran arrestos?


```python
%sql SELECT count(*) AS n_total_arrest FROM CHICAGO_CRIME_DATA WHERE LCASE(arrest)='true'
```

### Problema 4

##### ¿Qué clase de delitos de los registrados han sucedido en lugares del tipo GAS STATION (gasolinera)?


```python
%sql SELECT DISTINCT(PRIMARY_TYPE) AS gas_station_type_crimes FROM CHICAGO_CRIME_DATA WHERE LCASE("LOCATION_DESCRIPTION") = 'gas station'
```

Pista: ¿Que columna enumera, por ejemplo, los tipos de delitos como: THEFT (Robo)?


### Problema 5

##### En la tabla CENUS_DATA enumera todas las áreas cuyos nombres empiecen por la letra ‘B’


```python
%sql SELECT COMMUNITY_AREA_NAME from CENSUS_DATA GROUP BY COMMUNITY_AREA_NAME HAVING LCASE(COMMUNITY_AREA_NAME) LIKE 'b%'
```

### Problema 6

##### ¿Que escuelas desde el área 10 al 15 están certificadas como escuelas saludables?


```python
%%sql 

SELECT NAME_OF_SCHOOL AS HEALTHY_CERTIFIED_SCHOOLS, COMMUNITY_AREA_NUMBER 
FROM CHICAGO_PUBLIC_SCHOOLS 
WHERE LCASE(HEALTHY_SCHOOL_CERTIFIED) = 'yes' AND COMMUNITY_AREA_NUMBER BETWEEN 10 AND 15
```

### Problema 7

##### ¿Cuál es la puntuación promedio de seguridad de los colegios?


```python
%%sql 

SELECT AVG(safety_score) AS avg_safety_score_school
FROM chicago_public_schools;
```

### Problema 8

##### Enumera las cinco primeras áreas por promedio de inscripción universitaria (por numero de estudiantes)


```python
%%sql

select Community_Area_Name, 
AVG(College_Enrollment) as AVG_ENROLLMENT 
from CHICAGO_PUBLIC_SCHOOLS 
group by Community_Area_Name 
order by AVG_ENROLLMENT desc 
limit 5 
```

### Problema 9

##### Utiliza una subconsulta para determinar cual área tiene el peor valor en la puntuación de seguridad


```python
%%sql 

select COMMUNITY_AREA_NAME, SAFETY_SCORE 
from chicago_public_schools
where SAFETY_SCORE = ( select MIN(SAFETY_SCORE) from chicago_public_schools ) 
```

### Problema 10

##### [Sin utilizar explícitamente un operador JOIN] Encuentra el Ingreso Per Cápita del Área con un nivel de seguridad en sus colegios de 1


```python
%%sql
select CD.community_area_number, CD.community_area_name, CD.per_capita_income 
from census_data CD, chicago_public_schools CPS 
where CD.community_area_number=CPS.community_area_number and CPS.safety_score = 1 
```

```python
%%sql 

select community_area_number, community_area_name, per_capita_income 
from census_data where community_area_number = ( select community_area_number from chicago_public_schools where safety_score =1 ) 
```

```python
%%sql 
SELECT per_capita_income, community_area_name
FROM CENSUS_DATA
WHERE LCASE(community_area_name) in
LCASE((SELECT community_area_name
FROM chicago_public_schools
GROUP BY name_of_school, community_area_name
HAVING AVG(safety_score) = 1))
```
